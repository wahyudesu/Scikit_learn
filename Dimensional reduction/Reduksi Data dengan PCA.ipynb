```python
#import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_wine
winedata = load_wine()
X, y = winedata['data'], winedata['target']
print(X) 
```

```python
import matplotlib.pyplot as plt

# Ubah lebar dan tinggi gambar
plt.figure(figsize=(3, 5))

# Mengurangi ukuran dot (contoh: 10)
plt.scatter(X[:, 1], X[:, 2], c=y, s=20)  # Ganti 10 dengan ukuran yang Anda inginkan

# Menambahkan judul plot
plt.title("Original Data")

plt.show()
```

```python
df = pd.read_csv("wine.csv")
inputs = df.iloc[:, 0:13]
labels = df.iloc[:,0]

inputs
```

```python
#Normalize data
from sklearn.preprocessing import StandardScaler
x1 = StandardScaler().fit_transform(inputs)
x1
```

```python
import matplotlib.pyplot as plt

# Ubah lebar dan tinggi gambar
figure = plt.figure(figsize=(3, 5))  # Ganti ukuran sesuai yang Anda inginkan

#ax = plt.gca()
plt.scatter(x1[:, 0], x1[:, 1], c=Labels, edgecolor='none')

plt.title("After Standarization")

plt.show()
```

```python
from sklearn.decomposition import PCA # import PCA

pca = PCA() # initialising a PCA instance.
wine = pca.fit_transform(x1) # The eigen-decomposition is done by using the fit() function; projections of the data in the PCA space is obtained using the transform() function. 
print(wine.shape)
```

```python
print(pca.explained_variance_)
print(pca.explained_variance_ratio_) 
```

```python
print(np.round(pca.explained_variance_,1))
print(np.round(pca.explained_variance_ratio_,1)) 
```

```python
var = np.sum(pca.explained_variance_[0:2])
print(var)
var_percentage = np.sum(pca.explained_variance_ratio_[0:2])*100
print(var_percentage,'%')
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
ax = plt.gca()
plt.plot(pca.explained_variance_, color='red', linestyle='dotted')
ax.set_title("Screen plot")
ax.set_xlabel("Index of principal components")
ax.set_ylabel("The explained varaince")
```

```python
figure = plt.figure(figsize=(3, 5))  # Ganti ukuran sesuai yang Anda inginkan

ax = plt.gca()
plt.scatter(wine[:,0],wine[:,1], c=Labels, edgecolor='none',)
ax.set_title("The PCA plot")
ax.set_xlabel("The first principle component")
ax.set_ylabel("The second principle component")
```
